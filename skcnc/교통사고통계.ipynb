```python
# (1) 사고유형분류가 "차대차"인 사고유형별 사망 비율을 막 그대로 시각화하시오.
# 단, 사망비율은 사망자수/사고건수로 계산하며, matplotlib에서 폰트는 아래 코드를 활용하여 맑은 고딕으로 설정합니다.
from matplotlib import pyplot as plt
plt.rcParams["font.family"] = 'Malgum Gothic'
```

```python
import pandas as pd
# 데이터 불러오기
df = pd.read_excel("accident_stat.xlsx")
df
```

```python
df2 = df.loc[df["사고유형대분류"] == "차대차"]
df2["사망비율"] =df2["사망자수"]/ df2["사고건수"]
plt.rcParams["font.family"] = 'Malgum Gothic'
plt.bar(range(len(df2)), df2["사망비율"])
plt.xticks(range(len(df2)), df2["사고유형"])
plt.show()
```

```python
# (2) 사고유형대분류별로 사망자수가 가장 많은 사고 유형을 출력하세요.
# 그 결과는 ['횡단중', '측면충돌', '공작물충돌']이어야 합니다.

df.loc[df.groupby("사고유형대분류")["사망자수"].idxmax()]
```

```python
df.loc[df.groupby("사고유형대분류")["사망자수"].idxmax(), "사망자수"]
```

```python
# (3) 사고유형별 증상비율을 계산하고 중상비율이 50%를 넘는 데이터만 중상비율.csv로 저장합니다.
# 이 csv파일은 다음과 같은 구조를 갖습니다.
df["중상비율"] = df["중상자수"]/df["사고건수"]
df.loc[df["중상비율"]>0.5].to_csv("중산비율.csv", index= False, encoding="cp949")
```
